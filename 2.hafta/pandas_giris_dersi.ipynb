# 4. Veri Temizleme ve Dönüştürme

## İlgili Sütunları Seçme

```python
# selected_columns = ['title', 'year', 'genre', 'duration', 'country', 'actors', 'avg_vote', 'votes']
df = df[['title', 'year', 'genre', 'duration', 'country', 'actors', 'avg_vote', 'votes']]
```

## Tekrar Eden Satırlar

```python
# duplicated().any(), drop_duplicates(subset=['col'], keep=False)
print(df.duplicated().any())
print(df.duplicated(subset=['year']).any())
```

## Veri Tipi Dönüştürme

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
```

## Null Değerler

```python
df = df.dropna(axis=1, how='any')
```

# 5. Groupby

<img src="imgs/groupby.png" alt="drawing" width="600"/>

```python
# get genre with highest mean rating (group.rat.mean.idx or nlargest)
df.groupby('genre').avg_vote.mean().nsmallest(10)
```

# 6. Veri Düzenleme

## Ekleme

<img src="imgs/adding_column.png" alt="drawing" width="600"/>

```python
# new column for voting sum
df['vote_sum'] = df['avg_vote'] * df['votes']
df
```

```python
# new row, df.loc[len(df.index)] = dict(cols: vals)
df.loc['müslüm'] = {'title':'Müslüm'}
df
```

## Silme

```python
# df.drop(label_list, axis) 0 for rows, 1 for columns
df = df.drop(['müslüm'], axis=0)
df
```

## Dönüştürme

```python
# df.apply(lambda x: func(x), axis)
df_copy = df.copy()
df_copy['title'] = df_copy['title'].apply(lambda x: x + " laz kit")
df_copy
```

# 7. Matematiksel İşlemler

```python
df1 = pd.DataFrame(np.random.randint(1, 10, size=(3, 3)), 
                   columns=list("ABC"), 
                   index=list("abc"))
df2 = pd.DataFrame(np.random.randint(1, 10, size=(3, 3)), 
                   columns=list("BCD"), 
                   index=list("bcd"))
print(df1)
print()
print(df2)
```

```python
# +, -, *, /, **, //
df1 + df2
```

```python
# np.exp etc.
np.exp(df1)
```

```python
# df.min(axis), df.max, df.sum, df.prod, df.mean, df.median ...
print(df1.min(axis=1))
print(df2.max(axis=0))
print(df1.mean())
```

# 8. Merging

<img src="imgs/merge.png" alt="drawing" width="600"/>

```python
df1 = pd.DataFrame({'ad': ['Bruce Banner', 'İsmail Abi', 'Sauron'],
                    'lakap': ['Hulk', 'Görüncek Adam', 'Dark Lord'],
                    'hobi': ['yoga', 'görünmek', 'yüzükler']})
df2 = pd.DataFrame({'ad': ['Bruce Banner', 'İsmail Abi', 'Sauron'],
                   'yas': ['52', '35', '15000'],
                   'memleket': ['Ohio', 'Istanbul', 'Orta Dünya']})
print(df1)
print()
print(df2)
print()
print(df1.merge(df2, on='ad'))
```
