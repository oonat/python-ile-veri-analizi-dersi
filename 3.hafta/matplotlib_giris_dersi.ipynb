# √ñrnek Kullanƒ±mlar

## Veri Seti

```python
df = pd.read_csv("./movies.csv")
df = df[['title', 'year', 'genre', 'duration', 'country', 'actors', 'avg_vote', 'votes']]
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df = df.dropna()
```

```python
df.head(5)
```

## Oylarƒ±n Analizi

```python
# Oy daƒüƒ±lƒ±mƒ± box plot
plt.boxplot(df.avg_vote)
plt.show()
```

```python
# Histogram
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df.avg_vote)
ax.grid(True)
bins = np.linspace(0, 10, 21)
ax.set_xticks(bins)

plt.show()
```

## Yƒ±llara G√∂re Analiz

```python
# yƒ±llara g√∂re film sayƒ±sƒ±
df_year_count = df.groupby('year').title.count()
plt.plot(df_year_count.index.to_numpy(), df_year_count.to_numpy())
plt.show()
```

```python
# yƒ±llara g√∂re avg vote
df_year_avg_vote = df.groupby('year').avg_vote.mean().nlargest(20)
plt.bar(df_year_avg_vote.index.to_numpy(), df_year_avg_vote.to_numpy())
plt.show()
```

## T√ºre G√∂re Analiz

```python
# t√ºre g√∂re film sayƒ±sƒ±
df_genre_count = df.groupby('genre').title.count().nlargest(10)
fig, ax = plt.subplots()
ax.pie(df_genre_count.to_numpy(), labels=df_genre_count.index.to_numpy(), autopct='%1.1f%%')
plt.show()
```

```python
# t√ºre g√∂re avg vote
df_genre_avg_vote = df.groupby('genre').avg_vote.mean().nlargest(10)
plt.bar(df_genre_avg_vote.index.to_numpy(), df_genre_avg_vote.to_numpy())
plt.xticks(rotation=90)
plt.show()
```

## Uzunluƒüa G√∂re Analiz

```python
# duration vs avg vote scatter
fig, ax_scatter = plt.subplots(figsize=(20, 10))
ax_scatter.scatter(df.duration, df.avg_vote)
plt.show()
```

## All-in-One

```python
fig, ax = plt.subplots(4, 2, figsize=(20, 20))

# box plot
ax[0][0].boxplot(df['avg_vote'])
ax[0][0].set_yticks(np.linspace(0, 10, 21)) 

# histogram
bins = np.linspace(0, 10, 21)
ax[0][1].hist(df['avg_vote'], bins=bins)
ax[0][1].set_xticks(bins)

# line graph
ax[1][0].plot(df_year_count.index.to_numpy(), df_year_count.to_numpy())  # Plot the chart

# bar chart
ax[1][1].bar(df_year_avg_vote.index.to_numpy(), df_year_avg_vote.to_numpy())

# pie chart
ax[2][0].pie(df_genre_count.to_numpy(), labels=df_genre_count.index.to_list(), autopct='%1.1f%%')

# bar chart
ax[2][1].bar(df_genre_avg_vote.index.to_list(), df_genre_avg_vote.to_numpy())
ax[2][1].tick_params(axis='x', labelrotation=90)

# scatter plot
ax[3][0].scatter(df.duration, df.avg_vote)

ax[3][1].text(0.5, 0.5, "Laz Kit Rulezz...", fontsize=36, horizontalalignment='center', verticalalignment='center')

fig.tight_layout()

plt.show()
```
