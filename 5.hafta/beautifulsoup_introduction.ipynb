# Beautifulsoup Kullanarak Web'ten Veri Çekmek

```python
import requests
from bs4 import BeautifulSoup
```

## 1. Web Sayfasına İstek Göndermek

```python
s = requests.Session()
header = {
    'User-Agent': "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:125.0) Gecko/20100101 Firefox/125.0",
}
s.headers.update(header)
```

```python
url = "https://eksisozluk.com/debe"
page = s.get(url)
```

```python
page.text
```

## 2. Beautifulsoup ile Text Parse Etmek

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
debe_list = soup.find(class_="topic-list partial")
```

```python
print(type(debe_list))
```

```python
debe_list.find_all("a")
```

```python
debe_items = []

for a in debe_list.find_all('a', href=True):
    url = a['href']
    caption = a.find(class_="caption").text
    debe_items.append((caption, url))
```

```python
print(debe_items)
```

```python
import time

debe_entries = []
eksi_root_url = "https://eksisozluk.com"

for caption, rel_url in debe_items:
    time.sleep(1) # Sleep for 1 seconds
    url = eksi_root_url + rel_url
    entry_page = s.get(url)
    soup = BeautifulSoup(entry_page.text, 'html.parser')
    entry_text = soup.find(class_="content").text
    author = soup.find(class_="entry-author").text
    entry_url = soup.find(class_="entry-date permalink")['href']
    entry_id = entry_url.split("/")[-1]
    entry_date = soup.find(class_="entry-date permalink").text
    debe_entries.append(
        (caption, entry_id, entry_date, author, entry_text))
```

```python
print(debe_entries)
```

## 3. Scraper Class

```python
import csv
import time

class EksiDebeScraper:
    def __init__(self, header):
        self.base_url = "https://eksisozluk.com"
        self.session = requests.Session()
        self.session.headers.update(header)
        
        self.debe_item_list = self._get_debe_table(f"{self.base_url}/debe")

    def _get_debe_table(self, debe_url):
        debe_page = self.session.get(debe_url)
        debe_parser = BeautifulSoup(debe_page.text, 'html.parser')
        debe_table = debe_parser.find(class_="topic-list partial")
                                                   
        debe_item_list = []
        for a in debe_table.find_all('a', href=True):
            url = a['href'].split("?")[0]
            title = a.find(class_="caption").text
            debe_item_list.append((title, url))
            
        return debe_item_list
    
    def _get_entry_content(self, title, entry_url):
        entry_page = self.session.get(entry_url)
        entry_parser = BeautifulSoup(entry_page.text, 'html.parser')
              
        entry_text = entry_parser.find(class_="content").text.strip()
        author = entry_parser.find(class_="entry-author").text
        entry_url = entry_parser.find(class_="entry-date permalink")['href']
        entry_id = entry_url.split("/")[-1]
        entry_date = entry_parser.find(class_="entry-date permalink").text
                                                   
        return title, entry_id, entry_date, author, entry_text
        
    def get_debe_list(self):
        return self.debe_item_list
                          
    def get_debe_entry(self, rank):
        title, rel_url = self.debe_item_list[rank]
        entry_url = self.base_url + rel_url
                                                   
        return self._get_entry_content(title, entry_url)                                           
        
    def save(self, fname):
        fields = ['title', 'entry_id', 'entry_date', 'author', 'entry_text']
        
        with open(fname, 'w') as csvfile:
            # creating a csv writer object
            csvwriter = csv.writer(csvfile)

            # writing the fields
            csvwriter.writerow(fields)

            for i in range(len(self.debe_item_list)):
                time.sleep(1)
                csvwriter.writerow(self.get_debe_entry(i))
```

```python
header = {
    'User-Agent': "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:125.0) Gecko/20100101 Firefox/125.0",
}

scraper = EksiDebeScraper(header)
```
